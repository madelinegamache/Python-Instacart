## Step 5  Data Quality & Consistency Checks

```python
# Check customers.csv data types
df_customers.dtypes
```

```python
# While user_id is a number, we won't use it for calculations so it should be changed to a "str" variable
df_customers['user_id'] = df_customers['user_id'].astype('str')
```

```python
# Check output
print(df_customers['user_id'].dtypes)
```

```python
# Checking the descriptive statistics of customers.csv
df_customers.describe()
```

```python
# Check for data consistency
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
  else:
    print ('No mixed dtypes', col)
```

```python
# Change the datatype of First Name to a str
df_customers['First Name'] = df_customers['First Name'].astype('str')
```

```python
# Check for data consistency again
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
  else:
    print ('No mixed dtypes', col)
```

```python
# Check for missing values again
df_customers.isnull().sum()
```

##### Now that the first name variable is a 'str' variable it doesn't register the blanks as missing values anymore.
