### Step 6

```python
# Creating “average_price” column using transform()
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking output
ords_prods_merge.head()
```

```python
# Create flags to assign "spending" labels to user ID based on mean_price
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequencies of spending flags for accuracy
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### Step 7

```python
# Creating “median_prior_orders” column using transform()
ords_prods_merge['median_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Creating a non-frequent order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating a regular customer order_frequency_flag
ords_prods_merge.loc[(ords_prods_merge['median_prior_orders'] > 10) & (ords_prods_merge['median_prior_orders'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Creating a frequent customer order_frequency_flag
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequencies of spending flags for accuracy
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Examining the null values
print(ords_prods_merge[ords_prods_merge.frequency_flag.isnull()])
```

#### The null values are all from one customer who has placed one order so the median can't be calculated. This customer will be categorized as a non-frequent customer.

```python
# Changing the null values in "frequency_flag" column to "Non-frequent customer"
ords_prods_merge['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
